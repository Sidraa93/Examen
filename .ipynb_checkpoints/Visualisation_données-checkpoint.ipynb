```python
!pip install scikit-learn
# Importer les modules nécessaires
import matplotlib.pyplot as plt  # Pour la visualisation
from sklearn.decomposition import PCA # Importer la classe PCA
from sklearn.datasets import make_classification



# Fonction pour générer des données avec des paramètres variables
def generate_data(n_samples=10000, n_features=20, n_informative=10, n_classes=3,
                  n_clusters_per_class=2, class_sep=1.0, random_state=42):
    X, y = make_classification(
        n_samples=n_samples,
        n_features=n_features,
        n_informative=n_informative,
        n_classes=n_classes,
        n_clusters_per_class=n_clusters_per_class,
        class_sep=class_sep,
        random_state=random_state
    )
    return X, y


X, y = generate_data()
print(f"Données générées : {X.shape[0]} échantillons, {X.shape[1]} caractéristiques")
```

```python
# Fonction pour visualiser les données avec PCA (2D)
def visualize_data(X, y, title="Visualisation des données (PCA)"):
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)

    plt.figure(figsize=(8, 6))
    for label in set(y):
        plt.scatter(X_pca[y == label, 0], X_pca[y == label, 1], label=f'Classe {label}')
    plt.title(title)
    plt.xlabel("Composante principale 1")
    plt.ylabel("Composante principale 2")
    plt.legend()
    plt.grid()
    plt.show()

# Tester avec des paramètres variables
params = [

    {'n_informative': 10, 'n_classes': 4, 'n_clusters_per_class': 2, 'class_sep': 4},
    {'n_informative': 15, 'n_classes': 2, 'n_clusters_per_class': 2, 'class_sep': 3},
    {'n_informative': 10, 'n_classes': 3, 'n_clusters_per_class': 1, 'class_sep': 2.5},

]

# Générer et visualiser les données pour chaque ensemble de paramètres
for param_set in params:
    X, y = generate_data(**param_set)
    visualize_data(X, y, title=f"Paramètres : {param_set}")
```
